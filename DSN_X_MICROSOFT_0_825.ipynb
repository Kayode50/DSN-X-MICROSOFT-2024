```python


def model_building(model,X_train,y_train,X_test,y_test,test_df):
  model.fit(X_train,y_train)
  y_pred = model.predict(X_test)


  acc_score = accuracy_score(y_test,y_pred)
  conf_mat = confusion_matrix(y_test,y_pred)
  #class_rep = classification_report(y_test,y_pred)
  test_pred = model.predict_proba(test_df)
  print(f'train score: {model.score(X_train,y_train)}')
  print(f'test score: {model.score(X_test,y_test)}')
  return acc_score,test_pred

models = [CatBoostClassifier(verbose=False),RandomForestClassifier(),KNeighborsClassifier(),SVC(probability=True),GaussianNB(),LogisticRegression(),xgb.XGBClassifier(),lgb.LGBMClassifier(verbosity=-1,class_weight='balanced')]
acc_scores = []
model_names = []
test_preds= []
for i in models:
  acc,test_pred = model_building(i,X_train,y_train,X_test,y_test,test_df)
  acc_scores.append(acc)
  test_preds.append(test_pred)

  model_names.append(str(i).split('(')[0])
  print(f'{i} : {acc}')
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_acc = []
pred_test_full = []
y_pred = np.zeros(len(x))
y_prob = np.zeros(len(x))
for train_index, test_index in skf.split(x, y):
    X_train, X_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    #resample = SMOTETomek(tomek=TomekLinks(sampling_strategy='majority'))
    X_train,y_train = NeighbourhoodCleaningRule(n_neighbors=3, threshold_cleaning=0.5).fit_resample(X_train,y_train)
    #X_train,y_train = resample.fit_resample(X_train,y_train)
    #X_train,y_train = SMOTE(random_state=42).fit_resample(X_train,y_train)
    #params = {'n_estimators': 228, 'max_depth': 9, 'learning_rate': 0.22413256337979365, 'bagging_temperature': 4.638400586684705, 'l2_leaf_reg': 2}
    #params = {'learning_rate': 0.014765380292123478, 'depth': 8, 'l2_leaf_reg': 3.6187836605095574, 'subsample': 0.6018480298825436, 'min_data_in_leaf': 28, 'bagging_temperature': 0.13114704108613529, 'random_strength': 2.6694894355591106}
    model = CatBoostClassifier(verbose=False,scale_pos_weight=4.4)
    #model = xgb.XGBClassifier()
    #model = RandomForestClassifier()
    model.fit(X_train,y_train)
    y_pred[test_index]= model.predict(X_test)
    y_prob[test_index] = model.predict_proba(X_test)[:,1]
    acc_score = accuracy_score(y_test,model.predict(X_test))
    cv_acc.append(acc_score)
    print(f'trainin score: {model.score(X_train,y_train)}')
    print(f'testing score: {model.score(X_test,y_test)}')
    pred = model.predict_proba(test_df)[:,1]
    #pred = model.predict(test_df)
    pred_test_full.append(pred)
    print(f'accuracy for this fold: {acc_score}')

print(f'average accuracy: {np.mean(cv_acc)}')
```

```python
def to_labels(pos_probs, threshold):
	return (pos_probs >= threshold).astype('int')
```

```python
#THRESHOLD TUNNING GRIDSEARCH CV
thresholds = np.arange(0, 1, 0.001)

scores = [accuracy_score(y,to_labels(y_prob,t)) for t in thresholds]
ix = np.argmax(scores)
print('Threshold=%.3f, accuracy score =%.5f' % (thresholds[ix], scores[ix]))
```
